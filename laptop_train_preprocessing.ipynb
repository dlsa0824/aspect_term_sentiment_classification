### 把每筆資料對每個aspect切開，讓一筆資料只能有一個aspect，如果一筆資料有多個aspect，重複text形成一對多的格式

```python
# 創一個空的dataframe
laptop_train_1 = pd.DataFrame(columns=['text','aspect','polarity'])

# 把每筆資料的text 對 aspect terms 分開成 一個text對一個aspect、polarity
for i in tqdm(range(len(laptop_train))):
    count = 0
    target_a = laptop_train.loc[i,'aspect_0'] # aspect_0值
    target_p = laptop_train.loc[i,'polarity_0'] # polarity_0的值
    while((count<=12) and (target_a != 'nan')): #如果count<=12 (因為dataframe只有0-12個aspect) 且 目標aspect值不是'nan'
        pd_temp = pd.DataFrame([[laptop_train.loc[i,'text'], target_a, target_p]],
                              columns=['text','aspect','polarity']) # 將text、aspect、polarity變成dataframe格式
        laptop_train_1 = laptop_train_1.append(pd_temp, ignore_index=True) # 將dataframe加在下面
        count += 1
        if count>12: # 判斷aspect數是否有超過資料集裡的aspect數，如果超過就break
            break
        a = 'aspect_'+ str(count) #目標放在下一個aspect column name
        p = 'polarity_'+str(count) # 下一個 polarity column name
        target_a = laptop_train.loc[i,a] # 下一個aspect 值
        target_p = laptop_train.loc[i,p] # 下一個polarity 值
    laptop_train.loc[i,'aspect count'] = str(count)
laptop_train_1
```

```python
# 看分開aspect term資料筆數有沒有一樣多
print('切成單一資料筆數:',len(laptop_train_1))
print('每個aspect相加:',len(aspect_1)+len(aspect_2)*2+len(aspect_3)*3+len(aspect_4)*4+len(aspect_5)*5+
      len(aspect_6)*6+len(aspect_7)*7+len(aspect_8)*8+len(aspect_13)*13)
```

```python
# 把polarity是conflict的資料刪除，(因為conflict資料量太少)
laptop_train_1=laptop_train_1.drop(laptop_train_1[laptop_train_1['polarity']=='conflict'].index)
print('扣掉label為conflict的資料數:',len(laptop_train_1))
laptop_train_1
```
